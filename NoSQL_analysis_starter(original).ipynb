### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}

# Use count_documents to display the number of documents in the result
print('Number of establishments with a hygiene score of 20:', establishments.count_documents(query))

# Display the first document in the results using pprint
if establishments.count_documents(query) > 0:
    pprint(establishments.find_one(query))
else:
    print("No establishments found with a hygiene score of 20")
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Define the query to find establishments in London with a rating value greater than or equal to 4
query_2 = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': 4}}

# Execute the query to find matching documents
result_2 = establishments.find(query_2)

# Count the number of documents in the result
count_result_2 = establishments.count_documents(query_2)

# Display the count of establishments
print('Count of London Establishments Rated 4 or Higher: ', count_result_2)

# Display the first document in the results using pprint if there are any
if count_result_2 > 0:
    pprint(result_2[0])
else:
    print("No establishments found matching the query.")



```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

# Define the latitude, longitude, and degree search
latitude = 51.490142
longitude = 0.08384
degree_search = 0.01

# Define the query
query = {
    'RatingValue': 5,
    'geocode.latitude': {'$gte': latitude - degree_search, '$lte': latitude + degree_search},
    'geocode.longitude': {'$gte': longitude - degree_search, '$lte': longitude + degree_search}
}

# Define the sort criteria
sort = [('scores.Hygiene', 1)]  # Sort by hygiene score in ascending order

# Retrieve the results
results = establishments.find(query).sort(sort)

# Print the results in a formatted list
for result in results:
    print(f"ID: {result['_id']}, Business Name: {result['BusinessName']}, Hygiene Score: {result['scores']['Hygiene']}")






```

```python
# Convert result to Pandas DataFrame


# Retrieve the results again to ensure the cursor is reset
results = establishments.find(query).sort(sort)

# Convert the results to a list and then to a DataFrame
result_df = pd.DataFrame(list(results))

# Display the DataFrame
print(result_df)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results
# Define the pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the pipeline and retrieve the results
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents:", len(result))

# Print the first 10 results
print("First 10 results:")
pprint(result[:10])
```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
# Convert the result to a Pandas DataFrame
result_df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print('Number of rows in the DataFrame:', len(result_df))

# Display the first 10 rows of the DataFrame
print('First 10 rows of the DataFrame:')
print(result_df.head(10))
```
